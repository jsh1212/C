## 문제 1

지난 수업 시간에는 numpy를 이용하여 sin 파 신호를 생성하는 방법과, 기본 주파수를 갖는 sin 파의 정수배 주파수를 가지는 배음을 혼합하면 음색이 달라지는 현상에 대해 학습하였다.

이를 바탕으로 아래 셀에 4개의 배음을 혼합한 소리를 생성하는 코드를 작성하고, 소리를 재생하여 음색을 확인하시오. 이때 각 배음의 진폭을 다양하게 조정해 가면서 가장 마음에 드는 음색을 도출하시오.

```python
# 소리 데이터에 필요한 외부 모듈 불러오기
import numpy as np                             # 행렬 및 벡터 데이터 관리를 위한 numpy 모듈
import matplotlib.pyplot as plt                # 소리 데이터의 그래프 표현을 위한 모듈
from scipy.io.wavfile import write              # wav 형식으로 소리 데이터를 저장하기 위한 모듈
import os                                      # wav 파일을 시스템 명령어로 재생하기 위한 모듈
from IPython.display import Audio, display     # colab 에서 소리 재생을 위한 모듈

# 환경 변수 설정하기
Fs = 44100.0                                   # 정보 샘플링 주기, 1초에 44100개의 샘플링, 단위는 Hz(주파수)

# 1초 데이터 생성을 위한 환경 변수 설정
tlen = 5                                       # 5초
Ts = 1/Fs                                      # 샘플링 사이의 간격(시간)
t = np.arange(0, tlen, Ts)                     # 소리 데이터를 생성할 시간 성분으로 구성된 배열로
                                               # [0, 1] 사이를 TimeStamp의 간격으로 분할하여
                                               # Fs 개의 데이터를 담을 수 있는 배열 t를 생성

sound = np.zeros_like(t)                       # 생성할 신호를 위한 배열 변수
                                               # 이 변수에 생성하는 음파를 저장할 것

################## 아래에 본인의 코드를 작성하시오. #######################
f0 = 440
harmonics = [1, 2, 3, 4]
amplitudes = [1.0, 0.5, 0.3, 0.2]

for i, harmonic in enumerate(harmonics):
    sound += amplitudes[i] * np.sin(2 * np.pi * f0 * harmonic * t)



##################  위에 본인의 코드를 작성하시오.  #######################

# sin 함수로 생성한 음성 데이터를 wav 형식의 파일로 저장
scaled_sound = np.int16(sound/np.max(np.abs(sound)) * 32767)
write('sound_cell1.wav', 44100, scaled_sound)

# 저장된 wav 파일 재생
display(Audio("sound_cell1.wav", autoplay=True))
```

## 문제 2

위 문제 1에서 생성한 음파는 4개의 배음을 혼합하였다. 이번에는 배음을 8개로 늘려서 음을 생성하시오. 이번에도 각 배음의 진폭을 다양하게 조정해 가면서 가장 마음에 드는 음색을 도출하시오.

```python
# 소리 데이터에 필요한 외부 모듈 불러오기
import numpy as np                             # 행렬 및 벡터 데이터 관리를 위한 numpy 모듈
import matplotlib.pyplot as plt                # 소리 데이터의 그래프 표현을 위한 모듈
from scipy.io.wavfile import write              # wav 형식으로 소리 데이터를 저장하기 위한 모듈
import os                                      # wav 파일을 시스템 명령어로 재생하기 위한 모듈
from IPython.display import Audio, display     # colab 에서 소리 재생을 위한 모듈

# 환경 변수 설정하기
Fs = 44100.0                                   # 정보 샘플링 주기, 1초에 44100개의 샘플링, 단위는 Hz(주파수)

# 1초 데이터 생성을 위한 환경 변수 설정
tlen = 5                                       # 5초
Ts = 1/Fs                                      # 샘플링 사이의 간격(시간)
t = np.arange(0, tlen, Ts)                     # 소리 데이터를 생성할 시간 성분으로 구성된 배열

sound = np.zeros_like(t)

# 기본 주파수와 배음 설정
f0 = 440  # 기본 주파수 (A4)
harmonics = [1, 2, 3, 4, 5, 6, 7, 8]  # 8개의 배음
amplitudes = [1.0, 0.5, 0.3, 0.2, 0.1, 0.05, 0.03, 0.01]  # 각 배음의 진폭

# 배음을 혼합
for i, harmonic in enumerate(harmonics):
    sound += amplitudes[i] * np.sin(2 * np.pi * f0 * harmonic * t)


sound8 = sound[:]

################## 아래에 본인의 코드를 작성하시오. #######################

additional_amplitudes = [0.1, 0.07, 0.05, 0.03, 0.01, 0.005, 0.002, 0.001]  # 진폭 조정
for i, harmonic in enumerate(harmonics):
    sound8 += additional_amplitudes[i] * np.sin(2 * np.pi * f0 * harmonic * t)

##################  위에 본인의 코드를 작성하시오.  #######################

# sin 함수로 생성한 음성 데이터를 wav 형식의 파일로 저장
scaled_sound = np.int16(sound8/np.max(np.abs(sound8)) * 32767)
write('sound_cell2.wav', 44100, scaled_sound)

# 저장된 wav 파일 재생
display(Audio("sound_cell2.wav", autoplay=True))
```

## 문제 3

위 문제 2에서 생성한 음파는 5초간 이어지는데, 이때 들리는 음의 처음과 끝이 동일한 세기를 가진다. 하지만 일반적으로 악기로 소리를 낼 때에는 처음에는 소리의 세기가 더 세다가 점차 세기가 줄어드는 형태로 소리를 낸다.

이번 문제에서는 문제 2에서 생성한 음파 신호의 후반부의 음파의 진폭이 자연스럽게 줄어들도록 신호를 가공하시오. 특정한 규칙을 따라야 하는 것은 없으며 최대한 자연스럽게 소리가 잦아드는 느낌이 들도록 가공해 보시오.

```python
# 소리 데이터에 필요한 외부 모듈 불러오기
import numpy as np                             # 행렬 및 벡터 데이터 관리를 위한 numpy 모듈
import matplotlib.pyplot as plt                # 소리 데이터의 그래프 표현을 위한 모듈
from scipy.io.wavfile import write              # wav 형식으로 소리 데이터를 저장하기 위한 모듈
import os                                      # wav 파일을 시스템 명령어로 재생하기 위한 모듈
from IPython.display import Audio, display     # colab 에서 소리 재생을 위한 모듈

# 환경 변수 설정하기
Fs = 44100.0                                   # 정보 샘플링 주기, 1초에 44100개의 샘플링, 단위는 Hz(주파수)

# 1초 데이터 생성을 위한 환경 변수 설정
tlen = 5                                       # 5초
Ts = 1/Fs                                      # 샘플링 사이의 간격(시간)
t = np.arange(0, tlen, Ts)                     # 소리 데이터를 생성할 시간 성분으로 구성된 배열

sound = np.zeros_like(t)


f0 = 440
harmonics = [1, 2, 3, 4, 5, 6, 7, 8]
amplitudes = [1.0, 0.5, 0.3, 0.2, 0.1, 0.05, 0.03, 0.01]


for i, harmonic in enumerate(harmonics):
    sound += amplitudes[i] * np.sin(2 * np.pi * f0 * harmonic * t)


sound_atten = sound[:]

################## 아래에 본인의 코드를 작성하시오. #######################


fade_out_duration = tlen / 2
fade_out_start = int(len(t) * (tlen - fade_out_duration) / tlen)


fade_out = np.linspace(1.0, 0.0, len(t) - fade_out_start)
sound_atten[fade_out_start:] *= fade_out

##################  위에 본인의 코드를 작성하시오.  #######################

# sin 함수로 생성한 음성 데이터를 wav 형식의 파일로 저장
scaled_sound = np.int16(sound_atten/np.max(np.abs(sound_atten)) * 32767)
write('sound_cell3.wav', 44100, scaled_sound)

# 저장된 wav 파일 재생
display(Audio("sound_cell3.wav", autoplay=True))
```
